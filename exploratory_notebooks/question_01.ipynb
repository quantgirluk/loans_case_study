Question 1: Consolidate the Static and Historic Data provided into a single pandas dataframe to allow for easier data exploration and facilitate future analysis.

The dataframe should be of the format where there is a row per loan per calendar month.

```python
import pandas as pd
```

Load the static data

```python
static = pd.read_excel('../data/2024_Strat_Casestudy.xlsx', sheet_name='DATA-Static', skiprows=2, usecols="B:I",)
```

```python
static.head()
```

Load the historic Balances

```python
historic_balances = pd.read_excel('../data/2024_Strat_Casestudy.xlsx', sheet_name='DATA-Month End Balances', index_col=0)
```

```python
historic_balances.head()
```

Reformat to a vertical shape by using the method `stack`

```python
balances = historic_balances.stack().reset_index(name='Balance').set_index(['loan_id', 'level_1'])
balances
```

Do the same for the Payment Due and Payment Made sheets

```python
historic_due = pd.read_excel('../data/2024_Strat_Casestudy.xlsx', sheet_name='DATA-Payment Due', index_col=0)
historic_due.head()
```

Note tha the name of the index is slightly different. So, we are going to rename it for consistency.

```python
historic_due.index.rename('loan_id', inplace=True)
historic_due.head()
```

```python
due_payments = historic_due.stack().reset_index(
    name='Payment_Due').set_index(['loan_id', 'level_1'])
```

```python
due_payments
```

```python
historic_made = pd.read_excel('../data/2024_Strat_Casestudy.xlsx',
                    sheet_name='DATA-Payment Made', index_col=0)
historic_made.head()
```

```python
made_payments = historic_made.stack().reset_index(
    name='Payment_Made').set_index(['loan_id', 'level_1'])
```

Next we put together the Balances and Payments

```python
all_historic_data = pd.concat([balances, due_payments, made_payments], axis=1).reset_index()
all_historic_data.head()
```

Finally, we merge with the static data by `loan_id`

```python
data = all_historic_data.merge(static, left_on=['loan_id'], right_on='loan_id')
data.head()
```
